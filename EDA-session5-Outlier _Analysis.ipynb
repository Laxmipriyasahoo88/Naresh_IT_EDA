```python
q1=np.percentile(visa_df['prevailing_wage'],25)
q1
```

```python
q2=np.percentile(visa_df['prevailing_wage'],50)
q2
```

```python
q3=np.percentile(visa_df['prevailing_wage'],75)
q3
```

```python
#IQR
IQR=q3-q1
IQR
```

```python
#calculate the lower and upper bound
lb=q1-1.5*IQR
ub=q3+1.5*IQR
lb,ub
```

```python
#find outlier
c1=visa_df['prevailing_wage']<lb
c2=visa_df['prevailing_wage']>ub
c=c1 | c2
outlier_df=visa_df[c]
outlier_df
```

```python
#nonoutliers
c1=visa_df['prevailing_wage']>lb
c2=visa_df['prevailing_wage']<ub
c=c1 & c2
non_outlier_df=visa_df[c]
non_outlier_df
```

```python
#compare original datt awith non outlier
plt.hist(visa_df['prevailing_wage'])
plt.hist(non_outlier_df['prevailing_wage'])
plt.show()
```
